```python
# set options to be headless, ..
from selenium import webdriver
import chromedriver_autoinstaller
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
chromedriver_autoinstaller.install()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
```

```python
wd = webdriver.Chrome(options=options)
```

```python
# load the webpage
wd.get('https://www.amazon.in')

wd.maximize_window()

# get the input elements
input_search = wd.find_element(By.ID, 'twotabsearchtextbox')
search_button = wd.find_element(By.XPATH, "(//input[@type='submit'])[1]")

# send the input to the webpage
input_search.send_keys("Smartphones under 10000")
sleep(1)
search_button.click()

# Start scrapping the data
products = []
prices = []
for i in range(10):
    print('Scraping page', i+1)
    product = wd.find_elements(By.XPATH, "//span[@class='a-size-medium a-color-base a-text-normal']")
    price = wd.find_elements(By.XPATH, "//span[@class='a-price'] | //span[@class='a-size-small'] | //*[contains(text(), 'No featured offers')]//parent::div/span[2] | //*[contains(text(), 'Price higher than typical')]//parent::div/span[2]")
    # print(prices)

    for p in product:
        products.append(p.text)
    for a in price:
        prices.append(a.text)
    next_button = wd.find_element(By.XPATH, "//a[text()='Next']")
    next_button.click()
    sleep(2)
```

```python
import pandas as pd

# Convert lists to DataFrame
data = {'Product Name': products, 'Price': prices}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
